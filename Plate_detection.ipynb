### Homework: Persian Car Plate Detection and Character Recognition Using YOLOv5

**Objective:**  
1. Create and label a dataset for detecting car plates.
2. Train a YOLOv5 model to detect the full car plate.
3. Crop the detected car plates and manually label the characters on them.
4. Train a YOLOv5 model to detect individual characters (digits/letters) on car plates.

---

### Part 1: Dataset Creation and Car Plate Detection

1. **Dataset Collection:**
    - Students must collect a set of images that feature cars with visible license plates. 
    - The dataset should contain diverse backgrounds and weather conditions to make the model robust.

2. **Labeling Car Plates:**
   - Using a labeling tool like **LabelImg** or **Label Studio**, students must create bounding box annotations for **entire car plates** in each image.
     - Ensure that the label for each bounding box is “plate”.

3. **Train YOLOv5 for Car Plate Detection:**
   - Install YOLOv5 and set up the training environment (you can provide the [YOLOv5 setup guide](https://github.com/ultralytics/yolov5) for reference).
   - Train the YOLOv5 model to detect the car plates from the labeled dataset.
     - Ensure they use the correct parameters and configurations for training, such as specifying the dataset, classes, and image size.
     - Evaluate the model’s performance using metrics like precision, recall, and mAP (mean average precision).

4. **Evaluation:**
   - After training, students should test the model on new, unseen images and check if it detects car plates accurately.
   - Students should report evaluation results, including any difficulties encountered (e.g., issues with angle, partial plates, blurry images).

---

### Part 2: Character Detection on Car Plates

1. **Crop Detected Car Plates:**
   - For each detected car plate, crop the plate from the original image using the bounding box coordinates.

2. **Labeling Individual Characters:**
   - Students will need to manually label the **individual characters** on each cropped car plate. This can be done using **LabelImg** or **Label Studio** as well.
     - The label for each bounding box should be the character (e.g., "A", "1", "X").
     - The goal is to have a separate bounding box for each character on the car plate (e.g., a license plate like “ABC 123” will have six bounding boxes: one for each letter/number).
     - Include multiple images with different characters (digits and letters) to train a robust character detection model.

3. **Training YOLOv5 for Character Detection:**
   - Using the labeled dataset of individual characters, train a YOLOv5 model specifically for detecting characters on the car plates.
     - Each image will now contain a cropped car plate, and each label will correspond to one character.
   - Evaluate the model to ensure that it correctly detects individual characters.

4. **Evaluation and Visual Results:**
   - After training, students should test the model on unseen car plates and visually verify that the characters are correctly detected (i.e., each digit/letter is enclosed in a bounding box).
   - Provide images that show both the plate and the detected characters with bounding boxes, along with the final evaluation metrics.
